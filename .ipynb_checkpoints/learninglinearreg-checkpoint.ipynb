```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"D:\prisha_manipal_sp\sp_rugved\co2.csv")
```

```python
def correlation_ratio(categories, values):
    values_mean = np.mean(values)
    categories_means = values.groupby(categories).mean()
    n = len(values)
    category_counts = values.groupby(categories).size()

    numerator = np.sum(category_counts * (categories_means - values_mean) ** 2)
    denominator = np.sum((values - values_mean) ** 2)

    return numerator / denominator

print(correlation_ratio(df['Fuel Type'], df['CO2 Emissions(g/km)']))
print(correlation_ratio(df['Make'], df['CO2 Emissions(g/km)']))
print(correlation_ratio(df['Vehicle Class'], df['CO2 Emissions(g/km)']))
print(correlation_ratio(df['Model'], df['CO2 Emissions(g/km)']))
print(correlation_ratio(df['Transmission'], df['CO2 Emissions(g/km)']))
```
